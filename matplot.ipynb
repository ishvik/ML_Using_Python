```python
import numpy as np
import matplotlib.pyplot as plt

t=np.arange(0.0,4.0,0.01)

s=1+np.cos(2*np.pi*t)

plt.plot(t,s)
plt.grid()

plt.xlabel('Time(t)')
plt.ylabel('Voltage(v)')
plt.title('cosine wave')

plt.show()
```

```python
t_time=np.arange(0.0,2.0,0.01)
s_vol=1+np.sin(2*np.pi*t_time)

plt.plot(t_time,s_vol)
plt.grid()

plt.xlabel('time(t)')
plt.ylabel('voltage(s)')
plt.title('sine wave')

plt.show()
```

```python
t_time=np.arange(0.0,2.0,0.01)
s_vol=1+np.tan(2*np.pi*t_time)

plt.plot(t_time,s_vol,'red')
plt.grid()

plt.xlabel('time(t)')
plt.ylabel('voltage(s)')
plt.title('tan wave')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1=np.linspace(0.0,5.0)
x2=np.linspace(0.0,5.0)

y1=1+np.cos(2*np.pi*x1)
y2=1+np.sin(2*np.pi*x2)

plt.subplot(2,1,1)
plt.plot(x1,y1,'o-')

plt.xlabel('x1')
plt.ylabel('y1')

plt.title('plot-x1 and y1')

plt.subplot(2,1,2)
plt.plot(x2,y2,'*-')

plt.xlabel('x2')
plt.ylabel('y2')
plt.title('plot-x2 and y2')

plt.show()
```

```python
x1=np.linspace(0.0,5.0)
x2=np.linspace(0.0,5.0)
x3=np.linspace(0.0,5.0)

y1=1+np.cos(2*np.pi*x1)
y2=1+np.sin(2*np.pi*x2)
y3=1+np.sin(2*np.pi*x3)

plt.subplot(3,1,1)
plt.plot(x1,y1,'o-')

plt.xlabel('x1')
plt.ylabel('y1')

plt.title('plot-x1 and y1')

plt.subplot(3,1,2)
plt.plot(x2,y2)

plt.xlabel('x2')
plt.ylabel('y2')
plt.title('plot-x2 and y2')

plt.subplot(3,1,3)
plt.plot(x3,y3)

plt.xlabel('x3')
plt.ylabel('y3')
plt.title('plot-x3 and y3')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(0,2*np.pi,0.01)

y1=np.sin(x)
y2=np.cos(x)

plt.plot(x,y1,label='1st')
plt.plot(x,y2,label='2nd')

plt.legend()
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```
